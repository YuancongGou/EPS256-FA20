## **Cross Correlation through a Matrix containing seismic traces**

```python
#input component
k=int(input("component: Z=2, N/S=1, E/W=0  "))
print ()
print()

if k==2: 
    print ("            Cross Correlation Coefficients for Z components, Episensors")
if k==1:  
    print ("            Cross Correlation Coefficients for N components, Episensors")  
if k==0:
    print ("            Cross Correlation Coefficients for E components, Episensors")  
print()
print ("          BK63      BK64     BK65      BK66     BK67     BK68      BK69   ")
print()

#setting up the cross correlation matrix CCMat
BK63Z = ev63[k].data
BK64Z = ev64[k].data
BK65Z = ev65[k].data
BK66Z = ev66[k].data
BK67Z = ev67[k].data
BK68Z = ev68[k].data
BK69Z = ev69[k].data
CCMat = [BK63Z,BK64Z,BK65Z,BK66Z,BK67Z,BK68Z,BK69Z]

#run two nested loops to cross correlate each row of CCMat with each other
#and generate print matrix Pmat of correlation coefficients at the same time


Pmat=np.zeros(64).reshape(8,8)


for i in range (7):
        for j in range (i+1):
            if j>6:
                break
            corr = correlate (CCMat[i],CCMat[j],200)
            Pmat[[i],[j]]=max(corr)
            Pmat[[j],[i]]=max(corr)
            
np.set_printoptions (precision=5)
print ("BK63   ",Pmat[0])
print ("BK64   ",Pmat[1])
print ("BK65   ",Pmat[2])
print ("BK66   ",Pmat[3])
print ("BK67   ",Pmat[4])
print ("BK68   ",Pmat[5])
print ("BK69   ",Pmat[6])
```
