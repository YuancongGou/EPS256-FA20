# **PySeis Modules**                                                  
>**Horst Rademacher, Berkeley Seismology Laboratory**
                                            (as of 27 Oct 20)

**Read SEED**

>Read, print file header and plot complete SEED file. Assumption: file contains three component data, hence three traces. Select time window for further analysis, plot seismogram of that time window and write a new file containing only the selected section of the original file. 

**Merge Files**

>Allows merging two continguous files into one new one containing data sets from both original files

**Read Files Q8**

>The Quanterra Q8 write data in unique mini-SEED files. Each is 4MByte long, with the followoing naming convention *BK-BK63-20200613151519.330* where BK is the network code, BK63 is the station name and the rest depicts the last the time at which the file was written. 

>This module reads those files and lists all traces containes therein. It then lets you select specific traces and plots them

**Plot Record Section**

>Plot a record section of seismograms sorted by distance. Also allows to plot with reduced velocity to highlight certain seismic phases

**Horiz Rot**

>Allows to rotate the N/S- and E/W-compoents into any direction within the horizontal plane, i.e. into radial (R) and transversal (T) directions. Currently does not allow 3D rotation including the incidience angle.

**Trace Overlay**

>Read the files containing the streams and traces you wish to compare

>Plot overlay traces for each component

>Includes a short sub-module for downsampling a data set.

**Seis Comparison**

>Variation of module Trace Overlay, without Donwsampling but with establishing a cross correlation matrix. 

**Diff Int**

>In this module I compute and plot the time derivative and the integral over time of a seismic time series. Goal: compare the recordings of accelerometers with broadband seismometers. This also lets you create displacement seismograms.

**Part Motion**

>Generate the particle motion in various planes from a displacement seismogram

**ACC Flip Test**

>This module allows the evaluation of a typical flip test of accelerometers. Print out the results in "g" for each component

**Calib Q8**

>In this module I can plot and evaluate the sine wave calibrations for broadband instruments. Plots overlays of injected signal over response signals of the three components

**Make PAZ (Poles and Zeroes)**

>Generate a set of poles and zeroes from two values: lower corner period and damping. Allows comparison plots of variations of either damping or of corner periods. 

**Transfer Fct**

>Use a given set of poles and zeroes to plot the amplitude transfer function, including determining the 3 dB point of the lower period. Does not yet contain plotting of the phase response. 

**Galperin Conv**

>In this module I convert the Z,N,E output of a seismometer with Galperin topology, also known as _homogeneous triaxial arrangement_ and used i.e. in the STS-2 series and the Nanometrics Trillium family, back to the original U,V,W traces of the three internal sensors. It also gives the option to re-convert from U,V,W to Z,N,E. This module contains only the conversion matrix for the STS-2 and not for the Trillium family of instruments. It also should not be used with the output of seismometers with a classical topology, where the three internal sensor are orthogonally arranged along the cardinal directions Z,N,E (i.e. STS-1, Guralp 3T or MBB-2).
